```python
import re
import pandas as pd
with open('test.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

for i in range(len(lines)):
    line=lines[i]
    if line[0]=='p':
        pattern = r'([^:]+):'
        match = re.search(pattern,line)
        if match:
            file = match.group(1).strip()
            filename=f'./test_csv/{file}.csv'
            data={'name':[],'x':[],'y':[],'p':[]}
    elif line[0]=='\t':
        pattern=r'\t([^:]+): X=(\d+), Y=(\d+), Score=(\d+\.\d+)'
        match = re.search(pattern,line)
        if match:
            name=match.group(1)
            x=int(match.group(2))
            y=int(match.group(3))
            p=float(match.group(4))
            data['name'].append(name)
            data['x'].append(x)
            data['y'].append(y)
            data['p'].append(p)
            if i==len(lines)-1: 
                df=pd.DataFrame(data)
                df.to_csv(filename,index=False)
            elif lines[i+1][0]=='p':
                df=pd.DataFrame(data)
                df.to_csv(filename,index=False)

    
        

    
```
